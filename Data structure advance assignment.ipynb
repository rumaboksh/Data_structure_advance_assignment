```python
### Q.6 Write a code to perform basic string compression using the counts of repeated characters.

ans) ##Here's a Python function to perform basic string compression using the counts of repeated characters:


def compress_string(s):
    if not s:
        return ""

    compressed = []
    count = 1
    for i in range(1, len(s)):
        if s[i] == s[i - 1]:
            count += 1
        else:
            compressed.append(s[i - 1] + str(count))
            count = 1

    compressed.append(s[-1] + str(count))
    compressed_str = "".join(compressed)

    return compressed_str if len(compressed_str) < len(s) else s

# Test the function
print(compress_string("aaabbbccc"))  # Output: "a3b3c3"
print(compress_string("abc"))  # Output: "abc"


##This function iterates through the input string, counting consecutive occurrences of each character. It builds a list of compressed characters and their counts, then joins them into a string. If the compressed string is shorter than the original, it returns the compressed string; otherwise, it returns the original string.
```

```python
### Q.20  Write a code to shuffle a given list randomly without using any built-in shuffle functions.

ans)  ##  Here is a simple code to shuffle a given list randomly without using any built-in shuffle functions:


import random

def shuffle_list(lst):
    for i in range(len(lst) - 1, 0, -1):
        j = random.randint(0, i)
        lst[i], lst[j] = lst[j], lst[i]
    return lst

# Test the function
lst = [1, 2, 3, 4, 5]
print(shuffle_list(lst))  # Output: [3, 5, 2, 1, 4] (randomly shuffled)


## This code works by iterating over the list from the last element to the first, and for each element, it swaps it with another randomly chosen element from the unshuffled part of the list.

## This algorithm is known as the Fisher-Yates shuffle, and it is a widely used and efficient method for shuffling a list randomly.

```

```python
### Q.27 Develop a code that takes a tuple of integers as input. The function should return the maximum and minimum values from the tuple using tuple unpacking.

ans)  ## Here is a simple code that takes a tuple of integers as input and returns the maximum and minimum values from the tuple using tuple unpacking:


def find_max_min(tup):
    *_, max_val = tup
    min_val = min(tup)
    return min_val, max_val

# Test the function
tup = (1, 2, 3, 4, 5)
min_val, max_val = find_max_min(tup)
print("Minimum value:", min_val)
print("Maximum value:", max_val)


## Here's how the code works:

# 1. The find_max_min function takes a tuple tup as input.

# 2. The line *_, max_val = tup unpacks the tuple, assigning all elements except the last one to the variable _ (which is a common convention for a "don't care" variable), and the last element to max_val. Since tuples are ordered, the last element is the maximum value.

# 3. The line min_val = min(tup) finds the minimum value in the tuple using the built-in min function.

# 4. The function returns the minimum and maximum values as a tuple.

# 5. In the test code, we call find_max_min with a sample tuple, unpack the result into min_val and max_val, and print the results.
```

```python
### Q.29 Write a code that takes a tuple and an element as input. The function should return the count of occurrences of the given element in the tuple.

ans) ## Here is a simple code that takes a tuple and an element as input and returns the count of occurrences of the given element in the tuple:


def count_element(tup, elem):
    return tup.count(elem)

# Test the function
tup = (1, 2, 2, 3, 2, 4, 2, 5)
elem = 2
print("Count of", elem, "in the tuple:", count_element(tup, elem))


## Here's how the code works:

# 1. The count_element function takes a tuple tup and an element elem as input.

# 2. The function returns the count of occurrences of elem in tup using the built-in count method of tuples.

# 3. In the test code, we create a sample tuple tup and an element elem, and then call count_element with these inputs.

# 4. The output will be the count of occurrences of elem in tup.



## Alternatively, we can use a loop to count the occurrences of the element:


def count_element(tup, elem):
    count = 0
    for x in tup:
        if x == elem:
            count += 1
    return count
```

```python
### Q.30 Develop a code that prompts the user to input two sets of strings. Then, print the symmetric difference of these two sets .

ans)  ##Here is a simple code that prompts the user to input two sets of strings and then prints the symmetric difference of these two sets:


def get_set_from_user(prompt):
    user_input = input(prompt)
    return set(user_input.split(','))

def main():
    print("Enter two sets of strings separated by commas.")
    set1 = get_set_from_user("Set 1: ")
    set2 = get_set_from_user("Set 2: ")
    symmetric_difference = set1 ^ set2
    print("Symmetric difference:", symmetric_difference)

if __name__ == "__main__":
    main()


## Here's how the code works:

# 1. The get_set_from_user function prompts the user to enter a set of strings, splits the input string into individual strings using the split method, and returns the result as a set.

# 2. The main function prompts the user to enter two sets of strings, gets the sets using the get_set_from_user function, finds the symmetric difference of the two sets using the ^ operator, and prints the result.



Also, the ^ operator returns a new set containing all elements that are in exactly one of the sets. It does not modify the original sets.

For example, if the user enters "a,b,c" for Set 1 and "c,d,e" for Set 2, # the output will be:


Symmetric difference: {'a', 'b', 'd', 'e'}
```

```python
### Q.31 Write a code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list.

ans) ## Here is a simple code that takes a list of words as input and returns a dictionary where the keys are unique words and the values are the frequencies of those words in the input list:


def word_frequencies(word_list):
    frequency_dict = {}
    for word in word_list:
        if word in frequency_dict:
            frequency_dict[word] += 1
        else:
            frequency_dict[word] = 1
    return frequency_dict

# Test the function
word_list = ["apple", "banana", "apple", "orange", "banana", "banana"]
print(word_frequencies(word_list))


## Here's how the code works:

# 1. The word_frequencies function takes a list of words word_list as input.

# 2. It initializes an empty dictionary frequency_dict to store the word frequencies.

# 3. It iterates over each word in word_list.

# 4. If the word is already in frequency_dict, it increments the count by 1. Otherwise, it adds the word to frequency_dict with a count of 1.

# 5. Finally, it returns frequency_dict.

When we run this code with the test input ["apple", "banana", "apple", "orange", "banana", "banana"],# it will output:


{'apple': 2, 'banana': 3, 'orange': 1}


##This shows that the word "apple" appears 2 times, "banana" appears 3 times, and "orange" appears 1 time in the input list.

## Alternatively, we can use the collections.Counter class to achieve the same result in a more concise way:


from collections import Counter

def word_frequencies(word_list):
    return Counter(word_list)
```

```python
### Q.32 Write a code that takes two dictionaries as input and merges them into a single dictionary. If there are common keys, the values should be added together.

ans)  ## Here is a simple code that takes two dictionaries as input and merges them into a single dictionary, adding the values together for common keys:


def merge_dictionaries(dict1, dict2):
    merged_dict = {**dict1, **dict2}
    for key in dict1:
        if key in dict2:
            merged_dict[key] = dict1[key] + dict2[key]
    return merged_dict

# Test the function
dict1 = {'a': 1, 'b': 2, 'c': 3}
dict2 = {'b': 3, 'c': 4, 'd': 5}
print(merge_dictionaries(dict1, dict2))


## Here's how the code works:

# 1. The merge_dictionaries function takes two dictionaries dict1 and dict2 as input.

# 2. It creates a new dictionary merged_dict by merging dict1 and dict2 using the {**dict1, **dict2} syntax.

# 3. It iterates over each key in dict1.

# 4. If the key is also in dict2, it adds the values together and updates the value in merged_dict.

# 5. Finally, it returns merged_dict.

When we run this code with the test input {'a': 1, 'b': 2, 'c': 3} and {'b': 3, 'c': 4, 'd': 5},# it will output:


{'a': 1, 'b': 5, 'c': 7, 'd': 5}


##This shows that the values for common keys 'b' and 'c' have been added together.

##Alternatively, we can use the collections.Counter class to achieve the same result in a more concise way:


from collections import Counter

def merge_dictionaries(dict1, dict2):
    return dict(Counter(dict1) + Counter(dict2))
```

```python
### Q.34 Write a code that takes a dictionary as input and returns a sorted version of it based on the values. You can choose whether to sort in ascending or descending order.

ans)  ## Here is a simple code that takes a dictionary as input and returns a sorted version of it based on the values in ascending order:


def sort_dict_by_values(input_dict):
    return dict(sorted(input_dict.items(), key=lambda item: item[1]))

# Test the function
input_dict = {'a': 3, 'b': 1, 'c': 2}
print(sort_dict_by_values(input_dict))  # Output: {'b': 1, 'c': 2, 'a': 3}


## Here's how the code works:

# 1. The sort_dict_by_values function takes a dictionary input_dict as input.

# 2. It uses the sorted function to sort the dictionary items based on the values.

# 3. The key argument of the sorted function is a lambda function that takes a dictionary item and returns its value.

# 4. The sorted function returns a list of sorted dictionary items.

# 5. The dict function is used to convert the list of sorted dictionary items back into a dictionary.

# 6. The function returns the sorted dictionary.

## If we want to sort in descending order, you can add the reverse=True argument to the sorted function:


return dict(sorted(input_dict.items(), key=lambda item: item[1], reverse=True))


##This will sort the dictionary items based on the values in descending order.
```
